# 🔧 Brightway 2.5 Setup Notebook (for UBC Jupyter Open)

This notebook installs the Brightway2.5 ecosystem in **UBC Jupyter Open**, where users can't create Conda environments directly.

✅ Use this notebook **only on Jupyter Open** (https://open.jupyter.ubc.ca)  
🛑 For local installs, use `brightway_env.yml` instead (see course README)

---

```python
# Quick diagnostics

import sys, shutil, platform
print("Current kernel Python:", sys.version.split()[0])
print("Current kernel executable:", sys.executable)
print("Conda available:", bool(shutil.which("conda")))
print("Platform:", platform.platform())

if not shutil.which("conda"):
    raise RuntimeError(
        "Conda was not found. On UBC Jupyter Open it should exist. "
        "Please open a Terminal and run `which conda`, or contact course staff."
    )

if sys.version_info >= (3, 13):
    print("\n⚠️ Note: This JupyterHub base kernel is Python 3.13. "
          "We will NOT install Brightway into it. "
          "This setup creates a separate Python 3.11 environment + kernel.\n")
```

```python
# Create the conda environment (in your HOME directory so it persists across JupyterHub restarts)
!test -x "$HOME/.conda/envs/bw25/bin/python" && echo "✅ Env bw25 already exists at $HOME/.conda/envs/bw25" || mamba create -y -q -p "$HOME/.conda/envs/bw25" python=3.11
```

```python
# Conda/mamba install (NO brightway2 meta-package)
!mamba install -y -q -p "$HOME/.conda/envs/bw25" -c conda-forge pandas numpy scipy matplotlib pypardiso seaborn
```

```python
# Remove older brightway2 meta-package if it exists (it causes conflicts with the new packages)
!conda run -p "$HOME/.conda/envs/bw25" python -m pip uninstall -y brightway2 || true
```

```python
# pip install Brightway 2.5 packages
!conda run -p "$HOME/.conda/envs/bw25" python -m pip install -U pip wheel setuptools -q
!conda run -p "$HOME/.conda/envs/bw25" python -m pip install -q \
  "bw2data>=4" "bw2calc>=2" "bw2io>=0.9" "bw_processing" \
  "bw2analyzer>=0.11" "stats_arrays>=0.7" "ecoinvent_interface>=2.4" ipykernel
```

```python
# Register the kernel (so it shows up in JupyterLab)
# (We remove any old bw25 kernelspec first to avoid stale paths like /opt/conda/envs/...)
!jupyter kernelspec remove -f bw25 > /dev/null 2>&1 || true
!conda run -p "$HOME/.conda/envs/bw25" python -m pip install -U ipykernel
!conda run -p "$HOME/.conda/envs/bw25" python -m ipykernel install --user --name bw25 --display-name "Brightway 2.5 (py311)"
```

```python
# Confirm installation

!conda run -p "$HOME/.conda/envs/bw25" python -c "import bw2data, bw2io, bw2calc; import sys; print('OK - imports work'); print('Python:', sys.version.split()[0]); print('bw2data:', bw2data.__version__); print('bw2io:', bw2io.__version__); print('bw2calc:', bw2calc.__version__)"
```
