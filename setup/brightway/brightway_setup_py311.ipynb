# 🔧 Brightway 2.5 Setup Notebook (for UBC Jupyter Open)

This notebook installs the Brightway2.5 ecosystem in **UBC Jupyter Open**, where users can't create Conda environments directly.

✅ Use this notebook **only on Jupyter Open** (https://open.jupyter.ubc.ca)  
🛑 For local installs, use `brightway_env.yml` instead (see course README)

---

```python
# Quick diagnostics

import sys, shutil, platform
print("Current kernel Python:", sys.version.split()[0])
print("Current kernel executable:", sys.executable)
print("Conda available:", bool(shutil.which("conda")))
print("Platform:", platform.platform())

if not shutil.which("conda"):
    raise RuntimeError(
        "Conda was not found. On UBC Jupyter Open it should exist. "
        "Please open a Terminal and run `which conda`, or contact course staff."
    )

if sys.version_info >= (3, 13):
    print("\n⚠️ Note: This JupyterHub base kernel is Python 3.13. "
          "We will NOT install Brightway into it. "
          "This setup creates a separate Python 3.11 environment + kernel.\n")
```

```python
# Create the conda environment

!conda env list | grep -q "^bw25" && echo "✅ Env bw25 already exists" || conda create -n bw25 -y python=3.11
```

```python
# Install Brightway packages via conda-forge

!conda install -n bw25 -y -c conda-forge brightway2 bw2data bw2calc bw2io bw_processing pandas numpy scipy matplotlib seaborn
```

```python
# Register the kernel (so it shows up in JupyterLab)
!conda run -n bw25 python -m pip install -U ipykernel
!conda run -n bw25 python -m ipykernel install --user --name bw25 --display-name "Brightway 2.5 (py311)"
```

```python
# Confirm installation

!conda run -n bw25 python -c "import bw2data, bw2io, bw2calc; import sys; print('OK - imports work'); print('Python:', sys.version.split()[0]); print('bw2data:', bw2data.__version__); print('bw2io:', bw2io.__version__); print('bw2calc:', bw2calc.__version__)"
```
